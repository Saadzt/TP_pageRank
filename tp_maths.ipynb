```python
import math
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[0, 0, 1, 1/2],[1/3, 0, 0, 0],[1/3, 1/2, 0, 1/2],[1/3, 1/2, 0, 0]])
n = len(A)
X0 = np.ones((n,1))/n
S= np.ones((n,n))*1/n
alpha = 0.15
gamma = 1/n
beta = 2*(gamma*n + 4)
M=(1-alpha)*A + alpha*S

def f(X):
    return (np.linalg.norm(np.dot(M,X)-X,2)**2) + gamma*(np.sum(X)-1)**2

def grad(A,X):
    M_X= np.dot(((1-alpha)*A),X) +  np.sum(X)*(alpha/n)
    MT_X=np.dot(((1-alpha)*A.T),X) +  np.sum(X)*(alpha/n)
    MT_M_X= np.dot(M.T,M_X)
    
    return  X - M_X - MT_X + MT_M_X + gamma*np.sum(X) - gamma

#question 1

# Descente de gradient 
mu= 1.99/beta
# D'apres un tp réalisé l'année derniere
x_star = [[0.368],[0.142],[0.288],[0.202]]
N=20
def descente_grad(A,X):
    L_x=[]
    L_fx=[]
    L_k=[]
    for k in range(0,N):
        X = X - 2*mu * grad(A,X)
        L_x.append(np.linalg.norm(X-x_star,2))
        L_fx.append(f(X))
        L_k.append(k)
    return X , L_x , L_fx , L_k

D = descente_grad(A,X0)[1]  #normes
B = descente_grad(A,X0)[2]  #f(xk)
C = descente_grad(A,X0)[3]  #[1,N]



plt.figure(figsize=(5,3))
plt.plot(C,D, label = "||xk-x*||")
plt.plot(C,B, label = "f(xk)")
plt.legend(bbox_to_anchor=(0.6, 0.6))
plt.title("descente gradient")
plt.show()
    

    




```

```python
#question5
from scipy import sparse

def sgd(A, X, y, learning_rate, num_iterations):

    A = sparse.csr_matrix(A)
    m = len(y)
    theta = np.zeros(A.shape[1])
    
    for i in range(num_iterations):
        # Tirage aléatoire d'un échantillon
        rand_index = np.random.randint(m)
        xi = A[rand_index:rand_index+1]
        yi = y[rand_index]
        
        # Mise à jour des paramètres
        error = yi - xi.dot(theta)
        theta = theta + learning_rate * error * xi.toarray()[0]
        
    return theta
```

```python


def descente_gradient_conjugue(A,X):
    L_x = []
    L_fx = []
    L_k = []
    r = -grad(A,X)
    N=10
    for k in range(0,N):
        grad_conj = grad(A,X)
        gamma_conj = -np.sum(grad_conj*r)/np.dot(np.dot(r.T,gamma + np.dot((np.identity(n) - M).T,(np.identity(n) - M))),r)
        X = X + gamma_conj*r
        grad_conj_ = grad(A,X)
        beta = np.linalg.norm(grad_conj_,2)**2/np.linalg.norm(grad_conj,2)**2
        r = grad_conj_ + beta*r 
        L_x.append(np.linalg.norm(X-x_star))
        L_fx.append(f(X))
        L_k.append(k)
    return X, L_x, L_fx, L_k
   

D = descente_gradient_conjugue(A,X0)[1]  #normes
B = descente_gradient_conjugue(A,X0)[3]  #[1,N]
C = descente_gradient_conjugue(A,X0)[2]  #f(xk)


plt.figure(figsize=(5,3))
plt.plot(B,D, label = "||xk-x*||")
plt.plot(B,C, label = "f(xk)")
plt.legend(bbox_to_anchor=(0.6, 0.6))
plt.title("descente gradient conjugué")
plt.show()
```
